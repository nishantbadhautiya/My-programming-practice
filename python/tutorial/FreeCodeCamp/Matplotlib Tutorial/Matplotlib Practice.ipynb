```python
import matplotlib.pyplot as plt

x = [1, 2, 3]
y = [2, 4, 6]
plt.plot(x, y)
plt.title("My First Graph", loc = 'center', y = 1, pad = 6.0, fontdict = {'fontsize': 18, 'color' : 'green', 'fontweight' : 800, 'fontname': 'Comic Sans MS'})
plt.xlabel("X axis", loc = 'center', fontdict = {'fontname' : 'Comic Sans MS', 'color' : 'red', 'fontweight': 800})
plt.ylabel("Y axis", loc = 'center', fontdict = {'fontname' : 'Comic Sans MS', 'color' : 'red', 'fontweight': 800})
plt.show()
```

```python
x = [0,1,2,3, 4]
y = [0,2,4,6, 8]
plt.title("My first Graph", fontdict = {'fontname' : 'Comic Sans MS', 'fontsize': 15, 'fontweight' : 600, 'color' : 'red'})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([1,2, 3,4])
plt.yticks([1,2,3,4,5,7.5,8])
plt.plot(x, y, label = '2x', color = 'green', linewidth = 2, marker = "o", markersize = 8, markeredgecolor = "yellow", markeredgewidth = 2, linestyle = "--")
plt.legend()
plt.show()
```
