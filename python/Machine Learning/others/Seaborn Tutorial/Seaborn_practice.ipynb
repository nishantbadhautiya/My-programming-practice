### Line Plot

```python
from matplotlib import pyplot as plt 
import numpy as np 
import pandas as pd 
import seaborn as sns 
```

```python
# using matplotlib
var_1 = [1,2,3,4,5,6]
var_2 = [1,2,3,1,7,6]
plt.plot(var_1, var_2)
plt.show()
```

```python
# using seaborn 
var_1 = [1,2,3,4,5,6]
var_2 = [1,2,3,1,7,6]
df = pd.DataFrame({"var_1" : var_1, "var_2" : var_2})
sns.lineplot(x = "var_1", y = "var_2", data=df)
plt.show()
```

```python
# using seaborn 
var_1 = [1,2,3,4,5,6]
var_2 = [1,2,3,1,7,6]
sns.lineplot(x=var_1, y=var_2)
plt.show()
```

```python
penguins = sns.load_dataset("penguins").head(50)
penguins.head(5)
```

```python
palettes = ['red', 'green']
markers = ['o', '^']
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data=penguins, hue="sex", size=1, style="sex",
            palette = palettes, dashes=False, markers=markers, legend="full")

plt.grid()
plt.title("Penguins Dataset")
plt.show()
```

```python
order = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x="island", y = "bill_length_mm", data=penguins, hue='sex', hue_order=['Female', 'Male'],
            order=order, errorbar=('ci', 100), n_boot=1000, dodge=True)

# ci Level of the confidence interval to show, in [0, 100].
# n_boot is Number of bootstrap samples used to compute confidence intervals.
plt.show()
```

```python
sns.set_theme(style='darkgrid')
order = ['Dream', 'Torgersen', 'Biscoe']
sns.barplot(x="island", y = "bill_length_mm", data=penguins, color='g', palette='hls', errcolor='r',
            errwidth=2, capsize=0.03, saturation=0.8, alpha = 0.8)
plt.show()
```
